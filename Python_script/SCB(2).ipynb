```python
# https://api-portal.sec.or.th/api-details#api=5a28f6df2b3a6d1788d2025c&operation=09-%E0%B8%9B%E0%B8%A3%E0%B8%B0%E0%B9%80%E0%B8%A0%E0%B8%97%E0%B8%81%E0%B8%AD%E0%B8%87%E0%B8%97%E0%B8%B8%E0%B8%99%E0%B8%95%E0%B8%B2%E0%B8%A1%E0%B8%A5%E0%B8%B1%E0%B8%81%E0%B8%A9%E0%B8%93%E0%B8%B0%E0%B8%9E%E0%B8%B4%E0%B9%80%E0%B8%A8%E0%B8%A9
from datetime import datetime
# fund_spec = []
rmf = []
ssf = []
now = datetime.now()
print(now)

# เนื่องจากบางกองทุนมี spec_code > 1 ฉะนั้นจึงต้องใช้ for each ไล่ check ในแต่ละ List
for i in sets:
    response2 = requests.get(f'https://api.sec.or.th/FundFactsheet/fund/{i}/specification', headers = headers2)
    if response2.status_code == 200:
        for j in json.loads(response2.text): # res.txt[0] , res.txt[1]
            if j['spec_code'] == 'SSF' :
                ssf.append(i)
                print(i)
                print(j)
            elif j['spec_code'] == 'RMF':
                rmf.append(i)
                print(i)
                print(j)

end = datetime.now()
print(end)
total = end - now
print(str(total)+ " secs")
```

```python
from datetime import date, timedelta
import time

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
start_date = date(2022, 1, 1)
end_date = date(2022, 12, 31)
invalid_days = 0
valid_days = 0
set_first_half_rmf = []
df = []
for obj_name in first_half_rmf:
    valid_days = 0
    invalid_days = 0
    print(obj_name)
    for single_date in daterange(start_date, end_date):
        response = requests.get(f'https://api.sec.or.th/FundDailyInfo/{obj_name}/dailynav/{single_date.strftime("%Y-%m-%d")}', headers = headers1)
        print(single_date.strftime("%Y-%m-%d"))
        print(response)
        print(response.text)
        if invalid_days >= 100 :
            break
        if response.status_code == 204: # กัน respone มาว่า no content
            print("Skipped")
            invalid_days += 1
            print("consecutive invalid days = ",invalid_days)
            continue
        invalid_days = 0
        valid_days += 1
        res = response.content
        XD = json.loads(res)
        time.sleep(0.1)
    if invalid_days >= 100 :
        useable = False
    else :
        useable = True
    set_first_half_rmf.append({'Name':obj_name,'Valid days':valid_days,'Useable':useable})
    df.clear()
```
