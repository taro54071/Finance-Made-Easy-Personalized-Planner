```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date

path = 'C:\\Users\\Thunyavut\\Desktop\\Finance-Made-Easy-Personalized-Planner\\Python_script\\ssf_nav_data\\'
file = 'M0017_2565_SSF.csv'
df = pd.read_csv(f'{path}{file}')
df
```

```python
X = df.iloc[:,2:3].values
y = df.iloc[:,1:2].values
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
X
```

```python
df.drop(columns=['net_asset', 'sell_price', 'buy_price', 'sell_swap_price', 'buy_swap_price'], inplace = True)
df
```

```python
df['nav_date'] = pd.to_datetime(df['nav_date'], format = '%Y-%m-%d')
df.rename(columns = {'nav_date':'ds', 'last_val':'y'}, inplace=True)
df
```

```python
train_size = int(len(X) * 0.7)
train = df[:train_size]
train
```

```python
test = df[train_size:]
print(len(test))
test
```

```python
from prophet import Prophet
m = Prophet(changepoint_prior_scale=0.03)
m.add_country_holidays(country_name='US')
m.fit(train)
```

```python
future = m.make_future_dataframe(periods=300, freq='d', include_history = False) 
future
```

```python
future = pd.merge(future, test, on='ds')
future = future.drop(columns='y')
future
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
fig = m.plot(forecast)
plt.scatter(test['ds'] , test['y'], c='red',s=7)
#plt.plot(forecast.iloc[-len(test):]['ds'], forecast.iloc[-len(test):]['yhat'], c='blue')
plt.legend(['Train data points','Predict','Uncertainty interval', 'Test data points','Predict'])
plt.show()
```

```python
merged_df = pd.merge(forecast, test, on='ds')
merged_df
```

```python
from statsmodels.tools.eval_measures import rmse

predictions = merged_df['yhat']
actuals = merged_df['y']

print(f"RMSE: {rmse(predictions, actuals)}")
merged_df[['yhat','y']]
```
